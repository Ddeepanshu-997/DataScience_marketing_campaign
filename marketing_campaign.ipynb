```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('marketing_campaign.csv', sep=';')
data
```

finding out the number of student that have particular degree

```python
pd=data['Education'].value_counts()
pd
```

making the graph of student vs degree 

```python
i1=data['Education'].value_counts().index
v1=data['Education'].value_counts().values
plt.bar(i1,v1,color=['r','b'])
```

finding out how many students are married

```python
ms=data['Marital_Status'].value_counts()
i2=data['Marital_Status'].value_counts().index
v2=data['Marital_Status'].value_counts().values
```

makimg the graph if it

```python
plt.bar(i2,v2 , color=['g','k'])
plt.yticks([100, 900], [200])
```

combining the graph

```python
plt.figure(figsize=(16,9), dpi=200)
plt.bar(i1,v1,color=['r','b'])
plt.bar(i2,v2 , color=['g','k'])
plt.xlabel('degree and maritial status',size=25)
plt.ylabel('degree holders with maritual status',size=25)
plt.legend()
plt.show()
```
